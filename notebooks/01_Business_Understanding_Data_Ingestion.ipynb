# Customer Churn Prediction using API-Driven Cloud Native Architecture

## 1. Business Understanding

Customer churn refers to customers leaving or discontinuing a service. In subscription-based businesses such as telecom, banking, SaaS, and digital services, churn directly affects revenue and customer lifetime value.

The objective of this project is to build a machine learning application that predicts whether a customer is likely to churn based on customer profile, service usage, account details, and billing information.

This project follows an API-driven cloud-native style architecture and includes:

- Data pipeline for ingestion, preprocessing, exploratory data analysis, and monitoring
- Machine learning pipeline for model training, evaluation, and metric logging
- API access using FastAPI to retrieve application details and prediction outputs
- DataOps workflow orchestration using Prefect
- Model monitoring and experiment tracking using MLflow

## 2. Project Objectives

The main objectives of this project are:

1. Develop an automated data pipeline for customer churn prediction.
2. Perform data preprocessing and exploratory data analysis.
3. Train and compare multiple machine learning models.
4. Select the best-performing model based on evaluation metrics.
5. Track model performance using MLflow.
6. Automate the workflow using Prefect.
7. Provide prediction and application information through FastAPI.
8. Demonstrate an API-driven cloud-native machine learning solution.

## 3. Dataset Description

This project uses the **Telco Customer Churn** dataset, which contains customer demographic information, subscribed services, account details, billing information, and churn status.

**Dataset Source**

**Target Variable**

  - Yes → Customer has left the company.
